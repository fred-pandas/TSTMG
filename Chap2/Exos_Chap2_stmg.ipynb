# Exo n°5

```python
import numpy as np

X = [0, 1, 2, 3, 4, 5, 6, 7]
Y = [613, 646, 673, 704, 719, 747, 777, 808]

a, b = np.polyfit(X, Y, 1)
a, b
```

Représentons le nuage de points et la droite de regression

```python
import pylab as pl

fig, ax = pl.subplots()
ax.plot(X, Y, 'o', color = 'blue')
ax.plot([0, 7], [a * 0 + b, a * 7 + b], color = 'red')
pl.show()
```
